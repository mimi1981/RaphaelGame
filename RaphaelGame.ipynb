```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def earnings(n,k):
    return 2**(n-2*k)
```

```python
def average_earnings(ntemp):
    num_earnings=0
    denom_earnings=0
    n=np.int(ntemp)
    for k in range(n+1):
        num_earnings=num_earnings+earnings(n,k)*math.comb(n, k)
        denom_earnings=denom_earnings+math.comb(n, k)
        
    return num_earnings/denom_earnings
        
```

```python
def std_earnings(ntemp):
    num_earnings=0
    denom_earnings=0
    n=np.int(ntemp)
    avg=average_earnings(n)
    for k in range(n+1):
        num_earnings=num_earnings+(earnings(n,k)-avg)**2.*math.comb(n, k)
        denom_earnings=denom_earnings+math.comb(n, k)
        
    return np.sqrt(num_earnings/denom_earnings)
```
